# Geopandas test
A short jupyter notebook to test installation of geopandas using pip and using Binder

```python
import geopandas as gpd
import contextily as cx
```

### Read shapefile & plot on default CRS

```python
shp_path = "./National_Parks_December_2022_GB_BFC_360652311650849397/NPARK_DEC_2022_GB_BFC.shp"
df = gpd.read_file(shp_path)
df.plot()
```

### Coordinate Reference System (CRS) for shapefile

```python
df.crs
```

### Plot shapefile on a map, including background tiles

```python
ax = df.plot(figsize=(10, 10), alpha=0.5, edgecolor='k')
cx.add_basemap(ax, crs=df.crs)
```
